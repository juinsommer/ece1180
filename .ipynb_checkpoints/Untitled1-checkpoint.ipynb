```python
from vpython import*
import numpy as np
import math
```

```python
class VSource:
    def __init__(self, vo, w):
        self.Vo = Vo # DC Gain (V)
        self.w = w # Frequency (Hz)
        
    def setFrequency(self, w):
        self.w = w
        
    def setAmplitude(self, Vo):
        self.Vo = Vo
        
    def getVoltage(self, t):
        Vi = self.Vo*sin(self.w*t)
        return Vi
```

```python
scene = canvas()

# Passive Components
c = 1e-6
r = 100

# Voltage Source
w = 40
Vo = 5
Vs = VSource(Vo, w)
Vt = 25e-3
n = 1
Is = 1e-9

# Time
t = 0
dt = .0001
dtn = .0001

Va = 0
Vb = 0
Ivi = 0

# Plot
d = []
time = []


while(t <= .7):    
    # Cacluate Vd
    Vd = Vb-Va
    
    # Calculate new Vi
    Vi = Vs.getVoltage(t)

    # Calculate new ICeq
    ICeq = -c/dt*Vb
    
    # Calculate Geq
    GDeq = Is/(n*Vt)*math.exp(Vd/(n*Vt))
    
    # Calculate Id
    Id = Is*(math.exp(Vd/(n*Vt))-1)
    
    # Calculate IDeq
    IDeq = Id - GDeq * Vd
    
    # Matrix
    A = np.matrix([[-GDeq, GDeq, 1], [GDeq, -GDeq-c/dt-1/r, 0], [1, 0, 0]])
    b = np.matrix([[IDeq], [ICeq-IDeq], [Vi]])
    x = np.linalg.solve(A, b)
    
    Va = x.item(0)
    Vb = x.item(1)
    Ivi = x.item(2)
    
    Vd_n = Vb-Va
    
    # if voltage difference is too large
    if(abs(Vd_n-(Vd)) >= .0001): # check for convergence
        if(dtn >= dt/10):
            dtn -= dt/10
            continue
      
    t += dtn
    dtn = dt

    d.append(abs(Vb))
    time.append(t)
    
graph()
g1 = gcurve(color=color.red)
for i in range(len(d)):
    g1.plot(time[i], d[i])
```
